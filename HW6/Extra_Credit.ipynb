```python
import matplotlib.pyplot as plt

# number of records
records = [50000, 100000, 150000]

# Point Queries
point_query_no_index = [15000, 31000, 93000]
point_query_index_branch_account_type = [32000, 62000, 32000]
point_query_index_balance = [0, 0, 0] #Execution Time With Index (e.g., balance) (µs)

# Range Queries
range_query_no_index = [16000, 31000, 344000]
range_query_index_branch_account_type = [31000, 109000, 188000]
range_query_index_balance = [15000, 32000, 218000]#Execution Time With Index (e.g., balance) (µs)

# Creating subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 10))  # one for point queries, one for range queries

# Plotting Point Queries
ax[0].plot(records, point_query_no_index, marker='o', linestyle='-', color='red', label='No Index')
ax[0].plot(records, point_query_index_branch_account_type, marker='o', linestyle='-', color='blue', label='Index on Branch Name & Account Type')
ax[0].plot(records, point_query_index_balance, marker='o', linestyle='-', color='green', label='Index on Balance (N/A)')

ax[0].set_title('Point Query Execution Times')
ax[0].set_xlabel('Number of Records')
ax[0].set_ylabel('Execution Time (µs)')
ax[0].legend()

# Plotting Range Queries
ax[1].plot(records, range_query_no_index, marker='o', linestyle='-', color='red', label='No Index')
ax[1].plot(records, range_query_index_branch_account_type, marker='o', linestyle='-', color='blue', label='Index on Branch Name & Account Type')
ax[1].plot(records, range_query_index_balance, marker='o', linestyle='-', color='green', label='Index on Balance')

ax[1].set_title('Range Query Execution Times')
ax[1].set_xlabel('Number of Records')
ax[1].set_ylabel('Execution Time (µs)')
ax[1].legend()

plt.tight_layout()
plt.show()
```
