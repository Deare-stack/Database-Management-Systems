```python
import matplotlib.pyplot as plt

# Data setup
records = [50000, 100000, 150000]  # Number of records
point_query_no_index = [15000, 31000, 93000]  # Point query times without index
point_query_index = [32000, 62000, 32000]  # Point query times with index
range_query_no_index = [16000, 31000, 344000]  # Range query times without index
range_query_index = [15000, 32000, 218000]  # Range query times with index

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(records, point_query_no_index, marker='o', linestyle='-', color='red', label='Point Query - No Index')
plt.plot(records, point_query_index, marker='o', linestyle='-', color='blue', label='Point Query - With Index')
plt.plot(records, range_query_no_index, marker='^', linestyle='-', color='green', label='Range Query - No Index')
plt.plot(records, range_query_index, marker='^', linestyle='-', color='purple', label='Range Query - With Index')

plt.title('Query Execution Times')
plt.xlabel('Number of Records')
plt.ylabel('Execution Time (Microseconds)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
